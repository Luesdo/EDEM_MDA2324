```python
ruta = "dataset_banco_clean.csv"
data = pd.read_csv(ruta)
```

```python
print(data.head())
```

```python
# Normalizar los datos numéricos
from sklearn.preprocessing import StandardScaler

# Seleccionar solo los datos numéricos
num_cols = data.select_dtypes(include=[np.number]).columns
data_num = data[num_cols]
scaler = StandardScaler()
scaler.fit(data_num)
data_num = scaler.transform(data_num)

# Generar dataframe
data[num_cols] = data_num
print(data.head())
```

```python
# Desviación típica = 1, media = 0
data.describe()
```

```python
# One hot encoding para las variables categóricas
# Seleccionar solo las variables categóricas
cat_cols = data.select_dtypes(include=[object]).columns
data_cat = data[cat_cols]

# One hot encoding
data_cat = pd.get_dummies(data_cat)

# Merge data_cat y data_num
data_clean = pd.get_dummies(data,columns=['job','marital','education','default','housing','loan','contact','poutcome','month'],drop_first=True)

print(data_clean.head())
```

```python
pd.DataFrame(data_clean)
```

## Dividimos datos y ponemos distintos modelos en práctica

```python
# Dividir los datos en entrenamiento, prueba y validación
from sklearn.model_selection import train_test_split

# Extraer la columna objetivo
y = data_clean["y"]
X = data_clean.drop(columns=["y"])

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Dividir los datos de prueba en prueba y validación
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=42)
```

```python
# Support Vector Machine
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Crear el modelo
model = SVC()

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir los datos de prueba
y_pred = model.predict(X_test)

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print("Precisión: {:.2f}".format(accuracy))
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

# Crear el modelo
model = LogisticRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir los datos de prueba
y_pred = model.predict(X_test)

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print("Precisión: {:.2f}".format(accuracy))
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

# Crear el modelo
model = RandomForestClassifier()

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir los datos de prueba
y_pred = model.predict(X_test)

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print("Precisión: {:.4f}".format(accuracy))
```

```python
# Voting Classifier
from sklearn.ensemble import VotingClassifier

# Crear los modelos
model1 = LogisticRegression()
model2 = RandomForestClassifier()
model3 = SVC()

# Crear el modelo de votación
model = VotingClassifier(estimators=[('lr', model1), ('rf', model2), ('svc', model3)], voting='hard')

# Entrenar el modelo
model.fit(X_train, y_train)

# Predecir los datos de prueba
y_pred = model.predict(X_test)

# Calcular la precisión
accuracy = accuracy_score(y_test, y_pred)
print("Precisión: {:.4f}".format(accuracy))
```

# Conclusiones

Dada la normalización llevada a cabo, con standard scaler y one hot encoding para variables categóricas, el modelo que mejor rendimiento aparente ofrece es el ensemble Random Forest Classifier.
